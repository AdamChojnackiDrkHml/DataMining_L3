```python
from numpy import random
import numpy as np
from matplotlib import pyplot as plt

random.seed(1)
```

```python
X = random.normal(0, 1, 100)
eps = random.normal(0, 0.25, 100)
```

```python
Y = -1 + 0.5 * X + eps
```

$|Y| = 100$

$\beta_0 = -1$

$\beta_1 = 0.5$

```python
plt.scatter(X, Y)
plt.xlabel("X")
plt.ylabel("Y")
```

```python
import statsmodels.api as sm

x = sm.add_constant(X)
model = sm.OLS(Y, x)
results = model.fit()
results.summary()
```

```python
Y_hat = X * results.params[1] + results.params[0]
Y_pop = 0.5 * X - 1
plt.plot(X, Y_hat, label="Model")
plt.plot(X, Y_pop, label="Popoluation regression")
plt.scatter(X, Y)
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
```

```python
XSquared = X**2

x = sm.add_constant(XSquared)
model = sm.OLS(Y, x)
result = model.fit()
result.summary()
```
